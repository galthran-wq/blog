## Math

The quadratic formula:

$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

A short inline sum: $\sum_{i=1}^n i = \frac{n(n+1)}{2}$.


```python
import math

def quadratic_roots(a, b, c):
    disc = b*b - 4*a*c
    return ((-b - math.sqrt(disc)) / (2*a), (-b + math.sqrt(disc)) / (2*a))

quadratic_roots(1, -3, 2)
```

## Plot example
A quick Matplotlib plot.


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 200)
y = x**2 - 3*x + 2

plt.plot(x, y)
plt.title("y = x^2 - 3x + 2")
plt.grid(True)
plt.show()
```
